## EDGE

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import cv2
import numpy as np

img = cv2.imread('hanni_shortcut.jpg',cv2.IMREAD_GRAYSCALE)
kernel = np.array([[1,1,1],[1,-8,1],[1,1,1]])
print(kernel)
output = cv2.filter2D(img,-1,kernel)
cv2.imshow('edge',output)
key = cv2.waitKey(0)
if key & 0xFF == ord('q'):
    cv2.destroyAllWindows()
```

## IDENTIFY

```python
kernel = np.array([[0,0,0],[0,1,0],[0,0,0]])
output = cv2.filter2D(img,-1,kernel)
cv2.imshow('Identify',output)
key = cv2.waitKey(0)
if key & 0xFF == ord('q'):
    cv2.destroyAllWindows()
```

## Ridge 

```python
kernel = np.array([[0,-1,0],[-1,4,-1],[0,-1,0]])
output = cv2.filter2D(img,-1,kernel)
cv2.imshow('Ridge',output)
key = cv2.waitKey(0)
if key & 0xFF == ord('q'):
    cv2.destroyAllWindows()
```

## Edge Detection

```python
kernel = np.array([[-1,-1,-1],[-1,8,-1],[-1,-1,-1]])
output = cv2.filter2D(img,-1,kernel)
cv2.imshow('Edge Detection',output)
key = cv2.waitKey(0)
if key & 0xFF == ord('q'):
    cv2.destroyAllWindows()
```

## Sharpen

```python
kernel = np.array([[0,-1,0],[-1,5,-1],[0,-1,0]])
output = cv2.filter2D(img,-1,kernel)
cv2.imshow('Sharpen',output)
key = cv2.waitKey(0)
if key & 0xFF == ord('q'):
    cv2.destroyAllWindows()
```

## Box Blur

```python
kernel = np.array([[1,1,1],[1,1,1],[1,1,1]])
kernel = kernel / 9
output = cv2.filter2D(img,-1,kernel)
cv2.imshow('Box Blur',output)
key = cv2.waitKey(0)
if key & 0xFF == ord('q'):
    cv2.destroyAllWindows()
```

## Gaussian Blur 3x3

```python
kernel = np.array([[1,2,1],[2,4,2],[1,2,1]])
kernel = kernel / 16
output = cv2.filter2D(img,-1,kernel)
cv2.imshow('Gaussian Blur 3x3',output)
key = cv2.waitKey(0)
if key & 0xFF == ord('q'):
    cv2.destroyAllWindows()
```

## Gaussian Blur 5x5

```python
kernel = np.array([[1,4,6,4,1],[4,16,24,16,4],[6,24,36,24,6],[4,16,24,16,4],[1,4,6,4,1]])
kernel = kernel / 256
output = cv2.filter2D(img,-1,kernel)
cv2.imshow('Gaussian Blur 5x5',output)
key = cv2.waitKey(0)
if key & 0xFF == ord('q'):
    cv2.destroyAllWindows()
```
