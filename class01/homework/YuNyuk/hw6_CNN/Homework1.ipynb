```python
import cv2
import numpy as np

img = cv2.imread('joyul.jpeg', cv2.IMREAD_GRAYSCALE)
kernel = np.array([[1, 1, 1],
                   [1, -8, 1], 
                   [1, 1, 1]])
print(kernel)
output = cv2.filter2D(img, -1, kernel)
cv2.imshow('edge', output)
cv2.waitKey(0)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

#   
img = cv2.imread('joyul.jpeg', cv2.IMREAD_GRAYSCALE)

#  
kernel = np.array([[1, 1, 1],
                   [1, -8, 1], 
                   [1, 1, 1]])

#  
output = cv2.filter2D(img, -1, kernel)

#    
plt.figure(figsize=(10, 5))

#   
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(img, cmap='gray')
plt.axis('off')

#   
plt.subplot(1, 2, 2)
plt.title('Filtered Image')
plt.imshow(output, cmap='gray')
plt.axis('off')

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

#   
img = cv2.imread('joyul.jpeg', cv2.IMREAD_GRAYSCALE)

#  
kernel = np.array([[1, 2, 1],
                   [2, 4, 2], 
                   [1, 2, 1]]) / 16.0


#  
output = cv2.filter2D(img, -1, kernel)

#    
plt.figure(figsize=(10, 5))

#   
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(img, cmap='gray')
plt.axis('off')

#   
plt.subplot(1, 2, 2)
plt.title('Gaussian Image')
plt.imshow(output, cmap='gray')
plt.axis('off')

plt.show()
```
