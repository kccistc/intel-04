```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 이미지를 흑백으로 읽어들임
img = cv2.imread('joyul.jpeg', cv2.IMREAD_GRAYSCALE)

# 커널 정의
kernel = np.array([[1, 1, 1],
                   [1, -8, 1], 
                   [1, 1, 1]])

# 필터 적용
output = cv2.filter2D(img, -1, kernel)

# 이미지를 주피터 노트북에 표시
plt.figure(figsize=(10, 5))

# 원본 이미지 표시
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(img, cmap='gray')
plt.axis('off')

# 필터링된 이미지 표시
plt.subplot(1, 2, 2)
plt.title('Filtered Image')
plt.imshow(output, cmap='gray')
plt.axis('off')

plt.show()
```
