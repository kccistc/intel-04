## 실습2

```python
def create_array(n):
    arr = np.zeros((n,n), dtype = int)
    for j in range(n):
        for i in range(n):
            arr[j,i] = n*j+i+1
    return arr

n = int(input("배열의 크기: "))
arr = create_array(n)
print(arr)
```
